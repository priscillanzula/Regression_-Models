# Visulaization using seaborn
1. Categorical plots:

a. Barplot()- barplot([x, y, hue, data, order, hue_order, â€¦])

   A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean.

   A visualization of the group by action. 

   To use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
import seaborn as sns
```

```python
df= sns.load_dataset('tips')
df.head()
```

```python
df.columns
```

#### 1. Numerical columns:
size, total_bill,tip.
#### 2. Categorical columns:
sex, smoker,day, time.

```python
sns.barplot(data = df,x= 'sex', y='tip', hue= 'smoker', palette ='viridis')
plt.show()
```

- Males tip more than females overall , regardless of whether they smoke or not.
- For females, smokers tip slightly more than non smokers.

```python
# change the estimator from mean to standard deviation
sns.barplot(x ='sex', y ='total_bill', data = df, hue= 'smoker',
            palette ='plasma', estimator = np.std)
plt.show()
```

```python
sns.barplot(x ='day', y= 'total_bill', data=df, color='navy', estimator= np.std)
plt.show()
```

- Saturday has the highest average total bills
- Friday shows the most variation in bill amounts (longest error bars), meaning customers spent very different amounts on Fridays compared to other days.


```python
sns.barplot(x ='sex', y ='total_bill', data = df, 
            palette ='plasma', estimator = np.std)
```

 The average total_bill for the male is more as compared to the female.

c. Boxplot()- boxplot([x, y, hue, data, order, hue_order, â€¦])

 It shows the distribution of the quantitative data that represents the comparisons between variables.

 Boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution 
i.e. the dots indicating the presence of outliers.

```python
sns.boxplot(x ='day', y ='total_bill', data = df, hue ='smoker')
```

```python
sns.boxplot( x="day",y="total_bill",data=df,hue="smoker",order=["Sun", "Sat", "Fri", "Thur"],hue_order=["Yes", "No"], palette= 'Set1')  
            #the values you use for hue_order must exactly match the unique values in your dataframe column that you passed as hue.
            #In your case: hue="smoker"
plt.show()
```

- X takes the categorical column and y is a numerical column.
Hence we can see the total bill spent each day."hue" parameter is used to further add a categorical separation.

```python
sns.boxplot(x= 'day', y= 'total_bill', hue='sex', data=df, palette='magma')
```

```python
sns.boxplot(
    x="day",
    y="total_bill",
    data=df,
    hue="sex",
    showmeans=True,
    meanprops={
        "marker": "o",              # marker shape (use 'o' instead of 'circle')
        "markerfacecolor": "white", # fill color
        "markersize": 10,           # size of the marker
        "markeredgecolor": "purple"  # edge color
    }
)
```

d. Violinplot() - violinplot([x, y, hue, data, order, â€¦])

It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution. 

```python
sns.violinplot(x ='day', y ='total_bill', data = df, hue ='sex', split = True)
plt.show()
#hue is used to separate the data further using the sex category
#
setting split=True will draw half of a violin for each level. This can make it easier to directly compare the distributions.
```

- The wider sections of the violin = more observations around that bill value.

- The narrower sections = fewer observations.

- Violin plots are powerful because they combine boxplot info (median, IQR, whiskers) with distribution shape (density).

e. Stripplot() - stripplot([x, y, hue, data, order, â€¦])
  
It basically creates a scatter plot based on the category.


- With strip plot is that you can't really tell which points are stacked on top of each other and hence we use the jitter parameter to add some random noise.

- jitter parameter is used to add an amount of jitter (only along the categorical axis) which can be useful when you have many points and they overlap, so that it is easier to see the distribution.

- Setting split=True is used to draw separate strip plots based on the category specified by the hue parameter.

```python
sns.stripplot(x ='day', y ='total_bill', data = df, jitter = True, hue ='smoker', dodge = True)
plt.show()
```

##### Distribution by day:

- Thursday & Friday: bills are clustered around 10â€“20, with a few higher outliers (up to 40+).

- Saturday & Sunday: wider spread of bills (from ~5 to 50). More variability on weekends.

##### Smoker vs Non-smoker:

Both groups overlap heavily (smokers and non-smokers spend similar amounts).

But non-smokers (orange) appear slightly more concentrated in the mid-range (15â€“25), while smokers (blue) show more extreme highs (outliers).

2. Distribution plots

a. distplot() - sns.distplot(a,                # dataset (array, list, or Series)
                             
    bins=None,        # number of histogram bins
                             
    hist=True,        # whether to plot histogram
                             
    kde=True,         # whether to plot KDE curve

    rug=False,        # whether to plot rug plot
                             
    fit=None,         # fit a scipy distribution
                             
    color=None,       # color of elements
                             
    label=None,       # legend label
                             
    norm_hist=False   # normalize histogram
)


It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset. Syntax:

distplot(a[, bins, hist, kde, rug, fit, ...])

### Distplot() is updated to displot()

```python
sns.displot(df['total_bill'], kde = True, color ='blue', bins = 30)
```

```python
sns.displot(df['total_bill'], bins= 5,kde=True,rug=False,color= 'navy',label=None)
```
